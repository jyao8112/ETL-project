```python
lat_r=df['coordinates.latitude']
lon_r=df['coordinates.longitude']
lat_m=parkingmeter_df['LATITUDE']
lon_m=parkingmeter_df['LONGITUDE']

resturant_lat_lon=list(zip(lat_r, lon_r))
meter_lat_lon=list(zip(lat_m, lon_m))
```

```python
#######################################################
##Calculate the distantce between meter and resturant##
#######################################################
from math import sin, cos, sqrt, atan2, radians
from heapq import heappush, heappop
# approximate radius of earth in km

R = 6373.0
restaurant_meter=[]

for i in range(0,1000):
    lat1 = resturant_lat_lon[i][0]
    lon1 = resturant_lat_lon[i][1]
    
    distance_list=[]
    meterid=[]
    for j in range(0,len(meter_lat_lon)):
        lat2 = radians(meter_lat_lon[j][0])
        lon2 = radians(meter_lat_lon[j][1])

        dlon = lon2 - lon1
        dlat = lat2 - lat1

        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c
        heappush(distance_list, (-distance, meter_lat_lon[j], parkingmeter_df['OBJECTID'][j]))
        if len(distance_list) > 10:
            heappop(distance_list)
    restaurant_meter.append(distance_list)
    
```
